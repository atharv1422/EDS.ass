```python
import pandas as pd
import matplotlib.pyplot as plt
x=df['Summary']
y=df['Temperature (C)']
plt.plot(x,y,marker='s',color='g')
plt.title("WEATHER FORECAST")
plt.xlabel("Summary of the Weather")
plt.ylabel("Temperature per Summary")
plt.show()
```
