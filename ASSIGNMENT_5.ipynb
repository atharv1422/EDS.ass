```python
import matplotlib.pyplot as plt
import csv
x = []
y = []
with open('/content/Book1 Weather forecast.csv','r') as csvfile:
	lines = csv.reader(csvfile, delimiter=',')
	for row in lines:
		x.append(row[1])
		y.append(row[11])
plt.plot(x, y, color = 'g', linestyle = 'dashed',
		marker = 'o',label = "Weather Forecast Data")
plt.xticks(rotation = 90)
plt.xlabel('Summary of the Weather Data')
plt.ylabel('Daily Summary of the Weather Data')
plt.title('Weather Report', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```

```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
dataFrame = pd.read_csv("/content/Book1 Weather forecast.csv")
sb.set_theme(style="whitegrid")
sb.swarmplot(x = dataFrame["Pressure (millibars)"], y = dataFrame["Wind Speed (km/h)"], data=dataFrame)
plt.ylabel("Wind Speed")
plt.xlabel("Pressure")
plt.title("SWARN PLOT OF THE WEATHER FORECAST")
plt.show()
```
